### a. Value Counts Method

```python
import pandas as pd
```

```python
df_air = pd.read_csv('Air_Quality.csv')
df_air.head()
```

### 1. Value Counts

```python
# Counting name elements

# len function
len(df_air['Name'])
# .count() method
df_air['Name'].count()
```

```python
df_air['Name'].value_counts() # melihat isi data pada colom string
```

```python
df_air['Name'].value_counts(normalize=True) # divide by the sum values
```

```python
df_air['Geo_Place_Name'].value_counts(normalize=True)
```

### b. Sort DataFRame

```python
# Sort by column
df_air.sort_values('Data_Value')
```

```python
# Sort by a column
df_air.sort_values('Data_Value', ascending=False)
```

```python
# Sort descending by multiple columns
df_air.sort_values(['Data_Value','Geo_Join_ID'], ascending=False)
```

```python
# sort descending by multiple columns and update dataFRame(mengganti data FRame)
df_air.sort_values(['Data_Value','Geo_Join_ID'], ascending=False)
```

```python
# mengubah jadi huruf kecil semua jadi bisa pakai lamda juga
df_air.sort_values('Name', ascending=True, key=lambda col:col.str.lower())
```

### C. set_index() dan sort_index() methode

#### 1. create index

```python
import numpy as np
import random
```

```python
new_index = np.arange(0, 18862) # create number
```

```python
random.shuffle(new_index)
```

```python
new_index
```

```python
df_air['new_index'] = new_index # membuat kolom baru
df_air
```

#### 2. Set index

```python
df_air.set_index('new_index' , inplace=True) # menngganti index dengan kolom yang digunakan
```

#### 3. Sort Index

```python
# sort dataFrame by new index
df_air.sort_index()
```

```python
df_air.sort_index(ascending=False) # sort descending
```
